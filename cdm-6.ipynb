```python
import math
pi = math.pi

def PointsInCircum(r,n=100):
    return [(math.cos(2*pi/n*x)*r,math.sin(2*pi/n*x)*r) for x in range(0,n+1)]
```

```python
circle= PointsInCircum(2,n=60)
```

```python
dff = pd.DataFrame(circle)
dff.shape
```

```python
dff
```

```python
kmeansc = KMeans(n_clusters=2).fit(dff)
centroidsc = kmeansc.cluster_centers_
print(centroidsc)
```

```python
plt.scatter(dff[0], dff[1],c= kmeansc.labels_) 
plt.scatter(centroidsc[:,0], centroids[:,1], c='red')
plt.show()
```
