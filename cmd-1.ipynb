```python
# hadis haghshenas - 401112106
# project 1 - dade kavi mohasebati

# tabdil ax rangi be siah sefid 
# ba 2 ravesh : code amade + khodeman tabdil konim
# chap tasavir
# chap data RGB pixel ha

#***********************************************************************************
# farakhani ketabkhane(library) haye morede niaz
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np
from PIL import Image



# khandane tasvir + chap (bedune mehvar)
image = mpimg.imread('k3.png')
plt.axis('off')
plt.imshow(image)
plt.show()


#ابعاد پیکسلی تصویر   
print('Height of image:',(image.shape[0],'pixels'))
print('Width of image:',(image.shape[1],'pixels'))



# make a list with RGB data of picture
im = Image.open('k3.png')
pixels = list(im.getdata())



# taghire range ax ba code haye amade cv2 & plt
gray_img = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.imshow(cv2.cvtColor(gray_img , cv2.COLOR_BGR2RGB))
plt.show()

print('with python auto commands')



#*************************************************************************************
# tariif tabe baraye tabdile dasti rgb be grayscale
# majmu baze rgb bayad beine [0...1] ya [0...255] bashad 
# R & G ra 0 dar nazar gerefte va faghat B ra 1 migzarim
# mianigin (avg) in 3, be ma tasvire (1) ra midahad

def rgb_1_gray(img):
        grayImage = np.zeros(img.shape)
        R = np.array(img[:, :, 0])
        G = np.array(img[:, :, 1])
        B = np.array(img[:, :, 2])

        R = (R *0.0)
        G = (G *0.0)
        B = (B *1)

        Avg = (R+G+B)
        grayImage = img.copy()

        for i in range(3):
           grayImage[:,:,i] = Avg
           
        return grayImage   

    
# khandan ax k3 ba esme 'image' va dadan be tabe tarif shodeye :"rgb_1_gray"
image = mpimg.imread("k3.png")   
grayImage1 = rgb_1_gray(image)  
plt.axis('off')
plt.imshow(grayImage1)
plt.show()
print('tasvire tabe RGB_1_GRAY') 
  
    
    
# be elate tiregi ziad tasvir dobare tabe jadid tarif mikonim
# in bar RGB ra adad mosavi migzarim be tori ke sum=1 
# masalan R=0.3,  G=0.3,  B=0.4 
# tebghe output in tabe natije behtari midahad.

def rgb_2_gray(img):
        grayImage = np.zeros(img.shape)
        R = np.array(img[:, :, 0])
        G = np.array(img[:, :, 1])
        B = np.array(img[:, :, 2])

        R = (R * 0.3)
        G = (G * 0.3)
        B = (B * 0.4)

        Avg = (R+G+B)
        grayImage = img.copy()

        for i in range(3):
           grayImage[:,:,i] = Avg
           
        return grayImage       

    
# khandan mojadad axe 'k3' va dadan be tabe tarif shode (rgb_2_gray)
image = mpimg.imread("k3.png")   
grayImage2 = rgb_2_gray(image)  
plt.axis('off')
plt.imshow(grayImage2)
plt.show() 
print('tasvire tabe RGB_2_GRAY')



# print list RGB
print('             ')
print('             ')
print('             ')
print('             ')
print('here is the list of rgb data :')
print(pixels)
print(type(pixels))


# از آنجایی که دلیل ارگومان ثابت چهارم (255) را نمیدانم میتوانیم با حلقه زیر آن را حذف نماییم
tedad = len(pixels)
a=[]
p=pixels

for i in range (tedad):
    a = p[i]
    a = np.delete(a,3,0)
    p[i] = a  

    
print('             ')
print('             ')
print('             ')
print('             ')
new_pixels = p
print(new_pixels)



# THE END
```
