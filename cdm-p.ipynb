```python
dff_len= len(dff)
meghdar= range(dff_len)
dff= pd.DataFrame(dff, index=list (meghdar) , columns=dff.columns)
dff
```

```python
dff.iloc[:, :]
```

```python
sotune1=dff.iloc[:, 0]
sotune1
```

```python
sotune2=dff.iloc[:, 1]
sotune2
```

```python
sotune1=np.unique(sotune1)
sotune1
```

```python
sotune2=np.unique(sotune2)
print(len(sotune1))
print(len(sotune2))
```

```python
url_diff = np.setdiff1d(list(sotune1), list(np.intersect1d(sotune1,sotune2)))[0]
newlabels=['' for x in range(len(sotune1))]
labels= dict(zip(sotune1, newlabels ))
```

```python
G= nx.DiGraph()
```

```python
#import networkx as nx
G.add_nodes_from(sotune1)
edges=list(zip(sotune1 ,sotune2))
G.add_edges_from(edges)
nx.draw(G,with_labels=False, font_size=10,node_color='pink')
plt.show()
```

```python
nx.draw_networkx(G, labels=labels, node_color = 'green', arrows = True, node_size=0.1)
plt.figure(figsize=(15,15))
plt.show()
```

```python
page_ranks= nx.pagerank(G)
page_ranks= sorted(page_ranks.items())

for i in page_ranks:
    print(i[0] ,': ', i[1])
```
