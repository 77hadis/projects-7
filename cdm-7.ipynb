```python
# cluster=r=r0=10 
# tarife function purity

def purity(cluster, tlabel, plabel):
    
    docs = tlabel.size
    tedad_tlabel = np.unique(tlabel).size
    purity=0 
    
    for k in range(0,cluster): #sigma i=1 ...r
        nij=0
        nij_list=[]
        for j in range (0,tedad_tlabel): #baraye j ha: check kardane i ha
            for i in range(0,tlabel):
                if tlabel[i]==j and plabel[i]==k: nij=nij+1
            nij_list.append(nij)
        maxj= nij_list.sort.index
        purity= purity+ (maxj/docs)
        
    return purity
```
